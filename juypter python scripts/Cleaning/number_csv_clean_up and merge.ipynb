```python
import numpy as np
import pandas as pd
import ast
```

```python
numberdotcom_df = pd.read_csv("data_files/quick_merge_with_youtube.csv")    
movies_df = pd.read_csv("data_files/movies_metadata.csv")    

numberdotcom_df.columns
```

```python
# Convert string amounts to integer values
numberdotcom_df["Production Budget"] = numberdotcom_df["Production Budget"].replace({'\$': '', ',': ''},regex=True).astype("int64")
numberdotcom_df["Opening Weekend Revenue"] = numberdotcom_df["Opening Weekend Revenue"].replace({'\$': '', ',': ''},regex=True).astype("int64")
numberdotcom_df["Domestic Box Office"] = numberdotcom_df["Domestic Box Office"].replace({'\$': '', ',': ''},regex=True).astype("int64")
numberdotcom_df["Infl. Adj. Dom. Box Office"] = numberdotcom_df["Infl. Adj. Dom. Box Office"].replace({'\$': '', ',': ''},regex=True).astype("int64")
numberdotcom_df["International Box Office"] = numberdotcom_df["International Box Office"].replace({'\$': '', ',': ''},regex=True).astype("int64")
numberdotcom_df["Worldwide Box Office"] = numberdotcom_df["Worldwide Box Office"].replace({'\$': '', ',': ''},regex=True).astype("int64")
numberdotcom_df.dtypes
```

```python
# Adjusting movie Budget, International and Worldwide Box office to inflation
inflation_rate=numberdotcom_df["Infl. Adj. Dom. Box Office"]/numberdotcom_df["Domestic Box Office"]

numberdotcom_df["Infl. Adj. Production Budget"]=numberdotcom_df["Production Budget"]*inflation_rate
numberdotcom_df["Infl. Adj. Production Budget"]=numberdotcom_df["Infl. Adj. Production Budget"].astype("int64")
numberdotcom_df["Infl. Adj. Opening Weekend Revenue"]=numberdotcom_df["Opening Weekend Revenue"]*inflation_rate
numberdotcom_df["Infl. Adj. Opening Weekend Revenue"]=numberdotcom_df["Infl. Adj. Opening Weekend Revenue"].astype("int64")
numberdotcom_df["Infl. Adj. International Box Office"]=numberdotcom_df["International Box Office"]*inflation_rate
numberdotcom_df["Infl. Adj. International Box Office"]=numberdotcom_df["Infl. Adj. International Box Office"].astype("int64")
numberdotcom_df["Infl. Adj. Worldwide Box Office"]=numberdotcom_df["Worldwide Box Office"]*inflation_rate
numberdotcom_df["Infl. Adj. Worldwide Box Office"]=numberdotcom_df["Infl. Adj. Worldwide Box Office"].astype("int64")
numberdotcom_df.head()
```

```python
movies_df.head()
```

```python
genre=[]
titles_list=[]

length=len(movies_df['genres'])
movies_df.index
for i in range(length):
    try:
        genre.append(ast.literal_eval(movies_df.iloc[i]['genres'])[0]['name'])
        titles_list.append(movies_df.iloc[i]['title'])
    except:
        print(f'no values, {i}')
```

```python
genres_df=pd.DataFrame({'Title':titles_list, 'Genre': genre})
genres_df.drop_duplicates('Title',inplace=True)
```

```python
new_df=pd.merge(numberdotcom_df, genres_df,on='Title',how='left')
new_df.to_csv("data_files/quick_merge_with_youtube_VERSION2.csv",index=False,header=True)
```
